```python
df_upd['Customer_Age_upd'] = pd.cut(df['Customer_Age'], bins = [26, 40, 60, 73],
                                labels = ['Young', 'Middle-Aged', 'Old'])
df_upd
```

```python
df_upd.drop('Customer_Age', axis = 1, inplace = True)
```

```python
df_upd['Customer_Age_upd'].value_counts()
```

```python
df2 = df_upd[df_upd['Attrition_Flag'] == 'Attrited Customer']
df2['Customer_Age_upd'].value_counts().loc['Young']
```

```python
def create_df(i):
    dic_entry = {}
    percentage = []
    tot_tran_list = []
    attr_cust = df_upd[df_upd['Attrition_Flag'] == 'Attrited Customer']
    index = df_upd[i].value_counts().index
    value_num = df_upd[i].value_counts()
    for j in index:
        attr_num = value_num.loc[j]
        no_of_attr = len(attr_cust[attr_cust[i] == j])
        tot_trans_cost = df_upd[df_upd[i] == j]['Total_Trans_Amt'].sum()
        percentage.append(no_of_attr / attr_num)
        tot_tran_list.append(tot_trans_cost)
        
    dic_entry['no_under_category'] = df_upd[i].value_counts().values
    dic_entry['percent_attrited'] = percentage
    dic_entry['total_trans_cost'] = tot_tran_list
    new_df = pd.DataFrame(dic_entry, index = index)
    return new_df
create_df('Customer_Age_upd')

```

```python
create_df('Gender')
```
